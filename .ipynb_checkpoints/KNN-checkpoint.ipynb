```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("Resources/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
X = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag' ]]
X.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df['koi_disposition']
y.head()
```

```python
# Split data inot training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Look at data
X_train.head()
# y_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale data using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalize training data to be between 0 and 1
X_scaler = MinMaxScaler().fit(X_train)

X_train = X_scaler.transform(X_train)
X_test = X_scaler.transform(X_test)
```

# Train the Model



```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

```python
# Note that k: 11 provides the best accuracy where the classifier starts to stablize
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train, y_train)
print('k=11 Test Acc: %.3f' % knn.score(X_test, y_test))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# View parameter keys
knn.get_params().keys()
```

```python
# Create the GridSearchCV model using 'C' and 'tol' as the parameters
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': [3, 5, 10, 19],
              'weights': ['uniform', 'distance'],
            'metric': ['euclidean', 'manhattan']}
grid = GridSearchCV(
    knn,
    param_grid,
    verbose=3,
    cv = 3,
    n_jobs = -1
)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
# If I am reading this correctly, the score (80.4%) after tuning is hirer than the test model score (79.5%)
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)
print(predictions[:5])
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["blue", "red", "green"]))
```
